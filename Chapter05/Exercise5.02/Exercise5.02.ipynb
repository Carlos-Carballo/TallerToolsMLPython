# Exercise 5.02 - ReLU Neurons

In this exercise we will investigate the Rectified Linear Unit (ReLU) activation function and the effect the tunable weights have in modifying the output of ReLU neurons.

Import numpy and matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rc('text', usetex=True)
```

Define the ReLU activation function as a Python function

```python
def relu(x):
    return np.max((0, x))
```

Define the inputs ($x$) and tunable weights ($\Theta$) for the neuron.  In this example the inputs ($x$) will be 100 numbers linearly spaced between -5 and 5, initially $\Theta=1$.

```python
theta = 1
x = np.linspace(-5, 5, 100)
x
```

Compute the outputs ($y$)

```python
y = [relu(_x * theta) for _x in x]
```

Plot the output of the neuron vs the input

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111)

ax.plot(x, y)
ax.set_xlabel('$x$', fontsize=22);
ax.set_ylabel('$h(x\Theta)$', fontsize=22);
ax.spines['left'].set_position(('data', 0));
ax.spines['top'].set_visible(False);
ax.spines['right'].set_visible(False);
ax.tick_params(axis='both', which='major', labelsize=22)
```

Now set $\Theta=5$, recomupte and store the output of the neuron

```python
theta = 5
y_2 = [relu(_x * theta) for _x in x]
```

Now set $\Theta=0.2$, recomupte and store the output of the neuron

```python
theta = 0.2
y_3 = [relu(_x * theta) for _x in x]
```

Plot the three different output curves of the neuron ($\Theta=1, \Theta=5, \Theta=0.2$) on the one graph.

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111)

ax.plot(x, y, label='$\Theta=1$')
ax.plot(x, y_2, label='$\Theta=5$', linestyle=':')
ax.plot(x, y_3, label='$\Theta=0.2$', linestyle='--')
ax.set_xlabel('$x\Theta$', fontsize=22)
ax.set_ylabel('$h(x\Theta)$', fontsize=22)
ax.spines['left'].set_position(('data', 0))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(axis='both', which='major', labelsize=22)
ax.legend(fontsize=22)
plt.show()
```
