```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Seed_Data.csv')
df.head()
```

```python
df = df[['A', 'LK']]
df.head()
```

```python
eigenvectors, eigenvalues, _ = np.linalg.svd(df.values, full_matrices=False)
```

```python
eigenvalues
```

```python
eigenvalues = np.cumsum(eigenvalues)
eigenvalues
```

```python
eigenvalues /= eigenvalues.max()
eigenvalues
```
